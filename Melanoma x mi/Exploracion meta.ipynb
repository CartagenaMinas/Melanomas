```python
import os
os.listdir("Melanomas")
```

```python
PATH="Melanomas/data"
os.listdir(PATH)
```

```python
import pandas as pd
train=pd.read_csv(f"{PATH}/train.csv")
train.head()
```

```python
train.info()
```

```python
nans=train.isna().sum()
```

```python
train["sex"]=train["sex"].fillna("unk")
train["age_approx"]=train["age_approx"].fillna("unk")
train["anatom_site_general_challenge"]=train["anatom_site_general_challenge"].fillna("unk")
train.isna().sum()
```

```python
train["sex"].unique()
```

```python
test=pd.read_csv(f"{PATH}/test.csv")
test.head()
```

```python
test.isna().sum()
```

```python
test=test.fillna("unk")
test.isna().sum()
```

```python
train.to_csv(f"{PATH}/train_clean2.csv",index=False)
test.to_csv(f"{PATH}/test_clean2.csv",index=False)
```

```python
from sklearn.model_selection import train_test_split
train_split,val_split=train_test_split(train, 
                                       test_size=0.33, 
                                       shuffle=True, 
                                       random_state=42, 
                                       stratify=train["target"])
train_split.shape,val_split.shape
```

```python
train_split.to_csv(f"{PATH}/train_clean2_split.csv",index=False)
val_split.to_csv(f"{PATH}/val_clean2_split.csv",index=False)
```

```python
_,subset=train_test_split(train_split, 
                                       test_size=0.1, 
                                       shuffle=True, 
                                       random_state=42, 
                                       stratify=train_split["target"])
subset.shape
```

```python
subset.to_csv(f"{PATH}/subset_clean2.csv",index=False)
```
